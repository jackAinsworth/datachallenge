```python
def min_max_scale(arr):
    min_val = np.min(arr)
    max_val = np.max(arr)
    scaled_arr = (arr - min_val) / (max_val - min_val)
    return scaled_arr
```

```python
scaled_data1 = min_max_scale(data[:,0])
scaled_data2 = min_max_scale(data[:,1])
scaled_data_both = min_max_scale(data)

print(scaled_data1)
print(scaled_data2)
print(scaled_data_both)
```

```python
#plt.scatter(scaled_data_both[:,0],scaled_data_both[:,1])
plt.scatter(scaled_data1, scaled_data2)
```

```python
def standardize(arr):
    mean_val = np.mean(arr)
    std_val = np.std(arr)
    standardized_arr = (arr - mean_val) / std_val
    return standardized_arr
```

```python
std_data1 = standardize(data[:,0])
std_data2 = standardize(data[:,1])


print(std_data1)
print(std_data2)
```

```python
plt.scatter(std_data1, std_data2)
```
