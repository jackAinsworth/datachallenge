```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def generate_correlated_data(correlation, size=100):
    mean = [0, 0]
    cov = [[1, correlation], [correlation, 1]]
    x1, x2 = np.random.multivariate_normal(mean, cov, size).T
    return x1, x2
```

```python
fig, axes = plt.subplots(2, 2, figsize=(6, 6), sharex=True, 
sharey=True, squeeze=False)

np.random.seed(5934)
x1 = np.random.randn(100)
x2 = np.random.randn(100)

coef1 = np.arange(0.70, 0.75, 0.01)

for i in range(6):
    avg = x2 * coef1[i]
    
    #print(avg)
    
axes[0, 1].set_title("0,70 bis 0.75")
axes[0, 1].scatter(x1, avg)


coef2 = np.arange(-0.75, 0.70, 0.01)

for i in range(len(coef2)):
    avg = x2 * coef2[i]
    
    #print(avg)
    
axes[1, 0].set_title("-0.75 bis 0,75")
axes[1, 0].scatter(x1, avg) 
    

coef3 = np.arange(0.95, 1, 0.01)

arr = [6][100]

for x in x2:
    i =0
    for c in coef3:
        arr[i].append(x*c)
        i=i+1
        
        
print(len(arr))

    
    
    
axes[1, 1].set_title(">095")
axes[1, 1].scatter(x1, avg) 


axes[0, 0].set_title("base")
axes[0, 0].scatter(x1, x2)



plt.xlim(-4,4)
plt.ylim(-4,4)
plt.subplots_adjust(left=0.1, right=0.95, bottom=0.1, top=0.95, wspace=0.1, hspace=0.2)
```
