```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(0)
x = np.random.randint(1, 10, 30)
y = x+np.random.normal(0, 1, 30)
```

```python
x
```

```python
y
```

```python
np.argwhere(x==1).flatten()
```

## Ab hier startet dann die Musterl√∂sung

```python
m = []
up = []
for i in range(1,10):
    dummy = np.argwhere(x==i).flatten()
    m.append(y[dummy].mean())
    z=1.96*y[dummy].std()/np.sqrt(dummy.size)
    up.append(z)
m=np.asarray(m)
np.arange(1,10);
```

```python
ax = plt.plot(np.arange(1,10),m)
ax = plt.fill_between(np.arange(1,10), (m-up), (m+up), color='lightblue', alpha=0.6)
```
