```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.feature_selection import VarianceThreshold

# Daten laden
data = np.genfromtxt('zoo.csv')

# a) PCA durchführen
pca = PCA()
pca.fit(data)

# Scree-Plot erstellen
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_) + 1), pca.explained_variance_, marker='o', linestyle='-')
plt.title('Scree Plot')
plt.xlabel('Anzahl der Hauptkomponenten')
plt.ylabel('Eigenwerte')
plt.grid(True)
plt.show()

# Diagramm der aufsummierten Varianzen
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(pca.explained_variance_) + 1), np.cumsum(pca.explained_variance_ratio_), marker='o', linestyle='-')
plt.title('Kumulierte erklärt Varianz')
plt.xlabel('Anzahl der Hauptkomponenten')
plt.ylabel('Kumulierte erklärt Varianz')
plt.grid(True)
plt.show()




```

```python
# b) Merkmale mit geringer Varianz entfernen
selector = VarianceThreshold()
var_data = selector.fit_transform(data)




# Anzahl der Merkmale vor der Auswahl
print("Anzahl der Merkmale vor der Auswahl:", data.shape[1])

# Merkmale mit den drei kleinsten Varianzen entfernen
idx_low_variance_features = np.argsort(selector.variances_)[:3]
print("Anzahl der Merkmale", idx_low_variance_features)
```
